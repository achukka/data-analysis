```python
import numpy as np
import re
```

```python
retweet_name_re = re.compile('^RT @([A-Za-z]+[A-za-z0-9]+):')
```

```python
# Get user for the original tweet
def get_original_user_from_retweet(tweet):
    result = retweet_name_re.match(tweet)
    if result is None:
        return np.nan
    return result.groups()[0]
```

```python
# Apply the function to each column of the tweet(data frame)
tweets.tweet.apply(get_original_user_from_retweet)
```

```python
# Extract user name using the regular expression( extract method returns a dataframe)
# extact the column that matches the regular expression
tweets.tweet.str.extract(retweet_name_re)
```

```python
# Get the most influential user (i.e users with most number of followers)
groupbyname = users.groupby('name')
```

```python
groupbyname.groups
```

```python
# get most influential user from the grouping
def get_most_influential_user(grouping):
    result = []
    for name, dataFrame in grouping:
        count = len(dataFrame)
        followers = dataFrame.followers_count.sum()
        result.append({"name": name, "followers": followers, "count": count})
    return pd.DataFrame(result)
```

```python
get_most_influential_user(grouping=groupbyname)
```

```python
user_followers_cnt = get_most_influential_user(grouping=groupbyname)
```

```python
user_followers_cnt.sort_values('followers')
```

```python
user_followers_cnt.sort_values('followers').tail()
```

```python
user_followers_count = user_followers_cnt[ user_followers_cnt["count"] > 1]
```
